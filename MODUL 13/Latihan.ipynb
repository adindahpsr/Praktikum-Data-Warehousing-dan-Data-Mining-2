```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("tripadvisor.csv")
print(data)
```

```python
#Menghapus Kolom yang Tidak Diperlukan
X = data.drop(columns=['User ID'])
print(X)
```

```python
# Menampilkan Jumlah Fitur
num_features = X.shape[1]
print("Jumlah fitur input:", num_features, "fitur")
```

```python
#Implementasi PCA dengan Jumlah Fitur Awal
pca = PCA(n_components=num_features)
pca.fit(X)
```

```python
#Menampilkan Hasil Variance pada Tiap Principal Components
for i,j in enumerate(pca.explained_variance_ratio_):
    print("Fitur independen ke-", (i+1),
          "menghasilkan variance ratio sebesar",
         round(j,7))
```

```python
#Menampilkan Beberapa Principal Component Pertama Dengan 
#Cumulative Explained Ratio Minimal 90%
cummulative_variance = 0
num_pc = 0
for i,j in enumerate(pca.explained_variance_ratio_):
    if cummulative_variance < 0.9:
        num_pc += 1
        cummulative_variance += j
        print("Fitur independen ke-",
             (i+1),
             "menghasilkan variance ratio sebesar",
             round(j,7))
```

```python
#Implementasi PCA dengan Jumlah Fitur yang Dikurangi
pca_reduced = PCA(n_components=num_pc)
pca_reduced.fit(X)
```
