```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
#1
GroceryStore = pd.read_csv("/content/drive/MyDrive/dwdm/GroceryStoreDataSet.csv", names = ["product"], header=None)
GroceryStore
```

```python
records = list(GroceryStore["product"].apply(lambda x: x.split(",")))
print(records)
```

```python
association_rules = apriori(records, min_support=0.003,
                            min_confidence = 0.2,
                            min_lift = 3,
                            min_length = 2)

association_results = list(association_rules)
for i in association_results:
  print(i, "\n")
```

```python
print(association_results[0])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]

    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=========================================")
```
